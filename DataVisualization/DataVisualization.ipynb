Plot a box plot for distribution of age with respect to each gender along with the information about whether they survived or not. (Column names : 'sex' and 'age')</p></div>

```python
accent1 = {'0': '#95cca4', '1': '#0f99ca'}
numeric_cols = df.select_dtypes(include=['number']).columns
cat_cols = df.select_dtypes(include=['object', 'category']).columns 
```

```python
cols = numeric_cols.drop('Survived')
n_columns = len(cols)
fig,axes = plt.subplots(n_columns, 2,figsize=(10, 10),dpi=300)
for i, col in enumerate(cols):
    sns.kdeplot(data = df, x=col, hue=df['Survived'].astype(str), fill=True, ax=axes[i][0], palette=accent1)
    sns.boxplot(x=df[col], y=df['Survived'].astype(str), ax=axes[i][1], palette=accent1)
fig.tight_layout()
plt.title('Survived statistics')
plt.show()
```
