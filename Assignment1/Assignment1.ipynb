# Data Cleaning Using pd transforms

### Imputation

```python
df["Price"].fillna(df["Price"].mean(), inplace=True)
```

### Scaling

```python
numeric_features = df.select_dtypes(include=['int64', 'float64'])
mean = numeric_features.mean()
std = numeric_features.std()
```

```python
scaled_feat = (numeric_features-mean)/std
```

```python
df[numeric_features.columns]=scaled_feat
```

```python
df["City"]=df["CouncilArea"].apply(lambda x: x.split()[0])
```

```python
def pad_dates(x:str)->str:
  if len(x)==9:
    return "0"+x
  else:
    return x
df["Date"] = df["Date"].apply(pad_dates)
df['Date'] = pd.to_datetime(df['Date'], format='mixed', dayfirst=True)
```
