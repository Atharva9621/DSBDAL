```python
sns.set()
sns.set_palette('PiYG')
SNS_CMAP = 'PiYG'
```

```python
DATA_PATH = 'IRIS.csv'
df = pd.read_csv(DATA_PATH)
```

```python
fig, axes = plt.subplots(4, 2, figsize=(10, 20))
accent = {'Iris-setosa': 'crimson', 'Iris-versicolor': 'pink', 'Iris-virginica' : '#aacc88'}
for i, col in enumerate(['sepal_length', 'sepal_width', 'petal_length', 'petal_width']):
    sns.kdeplot(data = df, x=col, hue=df['species'], ax=axes[i, 0], fill=True, palette=accent)
    sns.boxplot(df, ax=axes[i, 1])
    axes[i, 1].axhline(y=df[col].mean())
    axes[i, 1].axhline(y=df[col].median(), color='green')
```

```python
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
sns.heatmap(df.select_dtypes(include=numerics).corr(), cmap='PiYG', annot=True, annot_kws={'fontsize':7}, fmt='.1g', vmin=-1, vmax=1, center= 0)
```
